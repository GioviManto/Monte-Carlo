**Monte Carlo** refers to the technique of
- Generating i.i.d. samples following a desired distribution;
- Estimating functionals of the desired distribution (mean, variance, density, etc) using the generated i.i.d. samples.

In the following code we use the Inverse Method to generate a Standard Univariate Gaussian

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Generate uniform random numbers in the range [0,1]
uniform_samples = np.random.uniform(0, 1, 1000)

# Apply the inverse CDF (quantile function) of the standard normal distribution
# This transforms the uniform samples to follow a standard normal distribution
normal_samples = stats.norm.ppf(uniform_samples)

# Plot the histogram of the resulting normal_samples to confirm they follow N(0,1)
plt.figure(figsize=(12, 6))

# Plot histogram of the transformed samples
plt.hist(normal_samples, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black')

# Plot the theoretical standard normal distribution for comparison
x = np.linspace(-4, 4, 1000)
plt.plot(x, stats.norm.pdf(x), color='darkblue', linestyle='--', linewidth=2, label='Theoretical N(0,1) PDF')

# Add labels and title
plt.xlabel("Value")
plt.ylabel("Density")
plt.title("Histogram of Samples Generated by Inverse Transform Sampling for N(0,1)")
plt.legend()
plt.show()
```

From here, we can generalize to a General Univariate Gaussian using the transformation: Z = μ + σY

If Y ∼ N(0,1) we use Z = μ + σY which Z ∼ N(μ, σ^2).
